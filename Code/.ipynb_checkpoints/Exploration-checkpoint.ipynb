## All imports

```python
import gzip
import re
import pandas as pd
import numpy as np
import distance
import random

import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
from pprint import pprint

%matplotlib inline
```

## Clean methods

```python
def retrievePositions(record):
    regex = "^([01])\|([01])"
    seq1,seq2 = [],[]
    p = re.compile(regex)
    for x in record.split(sep='\t'):
        m = p.match(x)
        if m:
            seq1.append(str(m.group(1)))
            seq2.append(str(m.group(2)))
    return seq1, seq2
```

```python
def file_len(path):
    with gzip.open(path, 'rb') as f:
        for i, _ in enumerate(f):
            pass
    return i + 1

def collectData(path):
    print("retrieving file size...")
    n = file_len(path)
    res1, res2 = [""]*n, [""]*n
    print("extracting data...")
    with gzip.open(path, 'rb') as f:
        i = 0
        for l in tqdm(f,total=n):
            s = l.decode('utf8')
            s1,s2 = retrievePositions(s)
            if len(s1)+len(s2) > 0:
                for j in range(len(s1)):
                    res1[j] += s1[j]
                    res2[j] += s2[j]
                i += 1
    print("completed.")
    return res1, res2
```

```python
def plotDistanceDistrib(dataset, f, fit=False):
    distances = list()
    for i,x in enumerate(dataset):#tqdm(enumerate(dataset), total=len(dataset)):
        for y in dataset[(i+1):]:
            distances.append(f(x,y,normalized=True))
    sns.distplot(distances, kde=fit)
    array = np.array(distances)
#     return np.unique(np.around(array, decimals=3), return_counts=True)
    return array
    
def plotSimilDistrib(dataset, fit=False):
    simil = list()
    for i,x in enumerate(dataset):#tqdm(enumerate(dataset), total=len(dataset)):
        for y in dataset[(i+1):]:
            d = distance.hamming(x,y)
            simil.append((len(x) + len(y) - d) * 1.0 / (len(x) + len(y) + d))
    sns.distplot(simil, kde=fit)
    array = np.array(simil)
#     return np.unique(np.around(array, decimals=3), return_counts=True)
    return array
```

```python
def distanceDistribRand(dataset, m):
    distances = np.zeros(m)
    n = len(dataset)
    for i in range(m):
        a, b = random.randint(0,n-1), random.randint(0,n-1)
        while a==b:
            a, b = random.randint(0,n-1), random.randint(0,n-1)
        distances[i] = distance.hamming(dataset[a], dataset[b], normalized=True)
#     return np.unique(np.around(distances, decimals=3), return_counts=True)
    return distances                                                                                            
```

```python
def similarityDistribRand(dataset, m):
    simil = np.zeros(m)
    n = len(dataset)
    for i in range(m):
        a, b = random.randint(0,n-1), random.randint(0,n-1)
        while a==b:
            a, b = random.randint(0,n-1), random.randint(0,n-1)
        x, y = dataset[a], dataset[b]
        d = distance.hamming(x, y)
        simil[i] = (len(x) + len(y) - d) * 1.0 / (len(x) + len(y) + d)
#     return np.unique(np.around(simil, decimals=3), return_counts=True)
    return simil
```

```python
def similarityWithFixed(dataset):
    n = len(dataset)
    simil = np.zeros(n)
    i = random.randint(0,n-1)
    ref = dataset[i]
    for j, seq in enumerate(dataset):
        d = distance.hamming(ref, seq)
        simil[j] = (len(ref) + len(seq) - d) * 1.0 / (len(ref) + len(seq) + d)
#     return np.unique(np.around(simil, decimals=3), return_counts=True), i
    return simil, i
```

```python
def buildDiploid(data0, data1):
    data = list()
    for seq0, seq1 in zip(data0, data1):
        data.append(''.join([str(a+b) for a,b in zip([int(c) for c in seq0], [int(c) for c in seq1])]))
    return data
```
